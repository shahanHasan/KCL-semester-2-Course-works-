```python
# 1_DiscriminantFunc_K-Means_widroHoffman

import numpy as np


class Widro_Hoff_Learning_Algorithm:

    def __init__(self, learning_rate=0.01, n_iters=12):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # init parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 0 else 0 for i in y])

        for _ in range(self.n_iters):
            
            for idx, x_i in enumerate(X):

                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)
                
                # Perceptron update rule
                update = self.lr * (y_[idx] - y_predicted)

                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

    def _unit_step_func(self, x):
        return np.where(x>=0, 1, 0)
```

```python
# # Quiz , Perceptron , Sequential Delta Learning algorithm.
# q2

import numpy as np

class Sequential_Delta_Learning_Algorithm:

    def __init__(self, Learning_rate = 0.10 , 
                 weight_initial_1 = 0.5, weight_initial_2 = 1.5,
                 weight_initial_3 = -2.5, weight_initial_4 = 2.5,
                 theta=-0.5 , 
                 iterations= 2, target_output= None):
        
        self.lr =  Learning_rate
        
        self.w1 = weight_initial_1
        self.w2 = weight_initial_2
        self.w3 = weight_initial_3
        self.w4 = weight_initial_4
        
        self.__theta= theta
        
        self.iter = iterations
        
        self.initial_weights = np.array([-theta , weight_initial_1, weight_initial_2, weight_initial_3, weight_initial_4])
        self.initial_weight = np.array([-theta , weight_initial_1, weight_initial_2, weight_initial_3, weight_initial_4])
        self.updated_weights = self.initial_weights
        
        self.t = target_output
        
        self.correctbeforelearning = self.correctSamplesBeforeLearning
        self.correctafterlearning = self.correctSamplesAfterLearning
        self.activationFunction = self.unit_Step_Activation_Function
        self.finalWeightAfterLearning = None


    def fit(self, x, t):
        # e.g : w = [-theta,w1,w2] x = [1,x1,x2,x3,x4]
        # Loop {  
        #   y(predict) = H(wx)
        #   update w 
        # }
        """
        t is the taget output
        we need feature vectors = x , weights = w , and t
        """
        #INIT params
        # x ,t 
        # t[i] - current target output/class
        # x[i] - current feature vector
        Number_of_samples , cols = x.shape
        print(f"initial weights : {self.updated_weights}")
        print(f"initial weights : {self.updated_weights.dtype}")
        
        
        
        for i in range(self.iter):
            """ Epoch Loop"""
            
            for j in range(Number_of_samples):
                print()
                print()
                print(f"------------ iter {j} ---------------")
                print()
                print()
                print(f"current feature vector : \n {x[j]}")
                print(f"current target output : \n {t[j]}")
            
           
                temp = np.insert(x[j], 0, 1)
                print(f"current feature vector Normalised: \n {temp}")
                  
                y_predicted = self.predict(self.updated_weights, temp)
                print(f"predicted class label : {y_predicted}")

                #Weight Update Rule
                # (t -y)
                error = t[j] - y_predicted
                change = (self.lr * error ) * temp
                self.updated_weights += change
                print(f"Updated weights : {self.updated_weights}")
            
        self.finalWeightAfterLearning = self.updated_weights
        print("----------------------------")
        print(f"Updated Weights :  {self.finalWeightAfterLearning}")
        print("----------------------------")
        print(f"Initial Weights : {self.initial_weight}")
        print("done")
        
        Percentage_Before = self.correctbeforelearning(self.initial_weight,Number_of_samples,x,t)
        print()
        print("------------------------------")
        print(f"% correctly classified before learning : {Percentage_Before}")
        print()
        
        Percentage_After = self.correctafterlearning(self.finalWeightAfterLearning,Number_of_samples,x,t)
        print()
        print("------------------------------")
        print(f"% correctly classified after learning : {Percentage_After}")
        print()
        


    def predict(self, w,x):
        # w and x are numpy arrays simulating matrix of weights and vectors
        Matrix_product = w @ x
        print(f"Matrix_product : {Matrix_product}")
        y_predicted = self.activationFunction(Matrix_product)
        return y_predicted



    def unit_Step_Activation_Function(self, x):
        if (x < 0):
#             print(0)
            return 0.0
        elif (x == 0):
            print(0.5)
            return 0.5
        else:
#             print(1)
            return 1.0

    def correctSamplesBeforeLearning(self,w_initial,No_of_samples,X,target):
        correctly_Predicted_Samples = 0
        for i in range(No_of_samples):
            temp = np.insert(X[i], 0, 1)
#             print(f"current feature vector Normalised: \n {temp}")
                  
            y_predicted = self.predict(w_initial, temp)
#             print(f"predicted class label : {y_predicted}")
            if (target[i] == y_predicted):
                correctly_Predicted_Samples += 1
            
        percentage = (correctly_Predicted_Samples / No_of_samples) * 100
        return percentage
    
    def correctSamplesAfterLearning(self,w_updated,No_of_samples,X,target):
        correctly_Predicted_Samples = 0
        for i in range(No_of_samples):
            temp = np.insert(X[i], 0, 1)
#             print(f"current feature vector Normalised: \n {temp}")
                  
            y_predicted = self.predict(w_updated, temp)
#             print(f"predicted class label : {y_predicted}")
            if (target[i] == y_predicted):
                correctly_Predicted_Samples += 1
            
        percentage = (correctly_Predicted_Samples / No_of_samples) * 100
        return percentage
            
```
