```python
SentimentTypes = set(df["Sentiment"])
print(f"Types of sentiments : {SentimentTypes}")
countNum = 0
sentimentDict = {}

for sentiment in SentimentTypes:
    #num = df["Sentiment"][df["Sentiment"] == sentiment].count()
    num = df['Sentiment'].loc[lambda sentimentInner: sentimentInner == sentiment].count()
    #print(f"numbers for sentiment({sentiment}) : {num}")
    countNum += num
    sentimentDict[sentiment] = num
#print(countNum)
#print(sorted(sentimentDict.items(), key = lambda kv:(kv[1], kv[0])))
SecondLastValue = list(sorted(sentimentDict.items(), key = lambda kv:(kv[1], kv[0])))[-2]
print(f"Second most popular sentiment : {SecondLastValue}")
print(type(SecondLastValue))
```

```python
"""
Task 1 Question 4
"""

WNL=WordNetLemmatizer()

def text_process(data):
    msg = [w.lower() for w in data if w.isalpha() or w.isspace()]
    msg = ''.join(msg)
    msg = [w for w in msg.split() if w not in stopwords]
    msg = [w for w in msg if len(w) > 2]
    msg = [WNL.lemmatize(word) for word in msg ]
    #msg = nltk.word_tokenize(data)
    return msg

newDataset = df.copy()
sample = newDataset["OriginalTweet"]
sample_target = newDataset["Sentiment"]

# Splitting the data - 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(sample , sample_target, test_size = 0.3, random_state = 99)
print(f"Training split input: {X_train.shape}")
print(f"Testing split input : {X_test.shape}")
print(f"Training split class: {y_train.shape}")
print(f"Testing split class : {y_test.shape}")

# Count Vector Train
CV=CountVectorizer(analyzer=text_process)
CV.fit(X_train)
```

```python
len(CV.vocabulary_)
```

```python
# Transform Train 
X_train=CV.transform(X_train)
# Transform Test
X_test = CV.transform(X_test)
```

```python
print(X_train.nnz)
print(X_test.nnz)
```

```python
# TFIDF
Tfidf=TfidfTransformer()
Tfidf.fit(X_train)
# Train and Test transform
Tfidf_val=Tfidf.transform(X_train)
X_test=Tfidf.transform(X_test)

print(X_test.shape)
```

```python

# Classifier
sentiment=MultinomialNB()
sentiment.fit(Tfidf_val,y_train)
result1=sentiment.predict(X_test)
```

```python
result1
```

```python


print(classification_report(y_test,result1))
```

```python
cf = confusion_matrix(y_test,result1 , labels= y_test.unique())
print(cf)
print(y_test.unique())
```

```python
cf = confusion_matrix(y_test, result1 , labels= y_test.unique())
plt.figure("Heat Map Confusion Matrix")
sns.heatmap(cf, annot=True)
plt.show()
```

```python
# Error rate and accuracy score for D one prime data set

cf = confusion_matrix(y_test,result1 , labels= y_test.unique())
print(cf)

# Class 0 

TP_0 = cf[0,0]
FP_0 = cf[0,1] + cf[0,2] + cf[0,3] + cf[0,4]
cf_0 = np.delete(cf, 0 , 1)
cf_0 = np.delete(cf_0, 0 , 0)
TN_0 = np.sum(cf_0)
FN_0 = cf[:,0].sum() - TP_0

# Class 1 

TP_1 = cf[1,1]
FP_1 = cf[1,0] + cf[1,2] + cf[1,3] + cf[1,4]
cf_1 = np.delete(cf, 1 , 1)
cf_1 = np.delete(cf_1, 1 , 0)
TN_1 = np.sum(cf_1)
FN_1 = cf[:,1].sum() - TP_1

# Class 2

TP_2 = cf[2,2]
FP_2 = cf[2,0] + cf[2,1] + cf[2,3] + cf[2,4]
cf_2 = np.delete(cf, 2 , 1)
cf_2 = np.delete(cf_2, 2 , 0)
TN_2 = np.sum(cf_2)
FN_2 = cf[:,2].sum() - TP_2

# Class 3

TP_3 = cf[3,3]
FP_3 = cf[3,0] + cf[3,1] + cf[3,2] + cf[3,4]
cf_3 = np.delete(cf, 3 , 1)
cf_3 = np.delete(cf_3, 3 , 0)
TN_3 = np.sum(cf_3)
FN_3 = cf[:,3].sum() - TP_2

# Class 4

TP_4 = cf[4,4]
FP_4 = cf[4,0] + cf[4,1] + cf[4,2] + cf[4,3]
cf_4 = np.delete(cf, 4 , 1)
cf_4 = np.delete(cf_4, 4 , 0)
TN_4 = np.sum(cf_4)
FN_4 = cf[:,4].sum() - TP_2

TP = TP_0 + TP_1 + TP_2 + TP_3 + TP_4
FP = FP_0 + FP_1 + FP_2 + FP_3 + FP_4
TN = TN_0 + TN_1 + TN_2 + TN_3 + TN_4
FN = FN_0 + FN_1 + FN_2 + FN_3 + FN_4

print(f"tn : {TN}")
print(f"fp : {FP}")
print(f"fn : {FN}")
print(f"tp : {TP}")

Error_Rate = (FP+FN) / (TN+FP+FN+TP)
print(f"Error Rate : {Error_Rate}")
print(f"Accuracy Score : {1-Error_Rate}")
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=cf,linewidths=.5, annot=True,square = True,  cmap = "twilight_shifted")
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
Error_Rate = round((Error_Rate*100), 2)
all_sample_title =  f'Error rate: {Error_Rate} %'
plt.title(all_sample_title, size = 15)
```
